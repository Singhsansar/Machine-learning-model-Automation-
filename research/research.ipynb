```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def save_and_return_correlation_matrix_image(dataframe, save_path):
    # Compute the correlation matrix
    correlation_matrix = dataframe.corr()

    # Display the heatmap and save it as an image
    sns.set(style="white")  # Set the style of seaborn
    plt.figure(figsize=(10, 8)) 
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Matrix Heatmap")
    plt.savefig(save_path, bbox_inches="tight")
    plt.show()

    return save_path

# Example usage:
# Assuming you have a DataFrame called 'my_dataframe'
# and you want to save the correlation matrix as an image file called 'correlation_matrix.png'
my_dataframe = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
image_save_path = 'correlation_matrix.csv'
resulting_image_path = save_and_return_correlation_matrix_image(my_dataframe, image_save_path)

# print(f"Correlation Matrix image saved at: {resulting_image_path}")
```
